c)

source-limited case

$ R_{star} $ : total photon count rate for star

$ R_{sky} $ : total photon count rate for star

$$\frac{S}{N} = \frac{R_{star}t}{\sqrt{R_{star}t + R_{sky}t}}$$

$$= \frac{t}{\sqrt{\frac{t}{R_{star}} + \frac{R_{sky}t}{R_{star}^{2}}}}$$

if $R_{star}>>R_{sky}$ then

$$\frac{S}{N} \sim \frac{t}{\sqrt{\frac{t}{R_{star}}}}$$

$$ \sim R_{star}^{1/2}t^{1/2}$$

$$\frac{S}{N} \propto t^{1/2}$$

background-limited case

$$\frac{S}{N} = \frac{R_{star}t}{\sqrt{R_{star}t + R_{sky}t}}$$

$$= \frac{(R_{star}/R_{sky})t}{\sqrt{(R_{star}/R_{sky})t+(1/R_{sky}^{2})t}}$$

if $R_{sky}>>R_{star}$ then

$$\frac{S}{N} \sim \frac{(R_{star}/R_{sky})t}{\sqrt{(R_{star}/R_{sky})t}} $$

$$ \sim (R_{star}/R_{sky})^{1/2}t^{1/2}$$

$$\frac{S}{N} \propto t^{1/2}$$
